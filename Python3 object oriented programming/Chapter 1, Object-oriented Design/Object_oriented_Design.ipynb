# Basic Concepts of object-oriented

### Object-oriented analysis (OOA):
the process of looking at a problem, system, or task, and identifying the objects and interactions between those objects.

### Output:
set of requirements.

---
### Object-oriented design (OOD):
the process of converting the requirements into an implementation specification.

### Output:
a set of classes and interfaces that could be implemented in any bject-oriented programming language.

---
### Object-oriented programming (OOP):
process of converting the design into a working program.

## What is the Diffrence between Association, aggregation, and composition?

### Association

 

*   It's the most basic relationship in OO, each object has his own independence.

*   There is no owner or parent class for these objects, each one has his own life cycle.


*   An association relationship can be represented as one-to-one, one-to-many, or many-to-many.
*   It Give a path on connection between two object, one can send message via , For example, a metod, and the other respond.





### Aggregation 



*   Kind of Association.

*   Each object has its  own life cycle with no ownership.
*   When object is a part of another, but the parent can exist without the child
*   It can be represented as one-to-one, one-to-many, or many-to-many.


#### EX:
The relation between Chess pieces and Chess Set is an Aggregation, Pieces is a part of chess set, but the parent class (chess set) can exist seperatly.



### Composition


*   It's a specialized form of aggregation

*   The child can't be existed seperatly from the parent, it will be destroyed when the parent is destroyed



#### EX:
The relation between Chess positions and Chess Board is a Composition, Chess positions can't exist seperatly from chess board.

### Inheritance


*   It's when one class inherits the attributes and methods of another class.

*   child class can have its own properties and methods.
*   class can  inherits only one parent (Single Inheritance) class or more than one (Multiple Inheritance).


*    Multi-level Inheritance: it's when class_1 is inherited by a class_2 and this class_2 also gets inherited by class_3 and so on...

## How can we diffrintiate between these relationships?



*   If there is a releation between two objects, where one send a request and receive a return, it's an assocoiation.
*   When an object is a part from another, but the child object can exist seperatly, it's an aggregation.

*   When an object is a part from another, but the child object can't exist seperatly, it's a Composition (Special form of Aggregation).
*   When an object is some shape of another, it's an Inheritance (BMW, Tesla, Ferrari, or Ford inherit Parent Car Class )
