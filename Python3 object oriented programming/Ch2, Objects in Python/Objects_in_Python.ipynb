### What's the diffrence between methods and normal functions?


All methods has a required argument: 

* It's conventionally named self, but we can call it any other name (GHonem, for example)
*  It is simply a reference to the object that the method is being invoked on.
* it enables us to  access attributes and methods of that object



````python
# we here invoked the object when we call the method, but python automatically do that for us
p = Point()
Point.reset(p)
print(p.x, p.y)
````

````python
# example of using a diffrent argument name than "self".

import math

class Point1:
    '''
    This will instansiate two attributes x, y with zero values using 
    move method.
    '''
    def reset(self):
        self.move(0, 0)

    def move(self, x, y):
        self.x = x
        self.y = y

    def calculate_distance(self, other_point):
        return math.sqrt(
        (self.x - other_point.x)**2 +
        (self.y - other_point.y)**2)


class Point2:
    '''
    This will instansiate two attributes x, y with zero values using 
    move method.
    '''
    def reset(GHonem):
        GHonem.move(0, 0)
        
    def move(GHonem, x, y):
        GHonem.x = x
        GHonem.y = y

    def calculate_distance(GHonem, other_point):
        return math.sqrt(
        (GHonem.x - other_point.x)**2 +
        (GHonem.y - other_point.y)**2)
````

````python
point1 = Point1()   # using self argument
point2 = Point2()   # # using GHonem argument

point1.reset()
point2.move(5,0)
print(point2.calculate_distance(point1))
print(point1.calculate_distance(point2))
````

We can notice that all things worked the same after changing self to "GHonem"

````python
point1.move(x = 6, y = 10)
print(point1.calculate_distance(point2))
print(point2.calculate_distance(point1))
````

### USing constructor __ init __

````python
import math

class Point:
    '''
    As we define x and y here, we set a default values, if we didn't define x
    and y
    '''
    def __init__(self, x = 33, y= 15):
        # we can use self.move(self, x = 33, y= 15) instead
        self.x = x
        self.y = y

    def move(self, x, y):
        self.x = x
        self.y = y

    def reset(self):
        self.move(0, 0)

    def calculate_distance(self, other_point):
        return math.sqrt(
        (self.x - other_point.x)**2 +
        (self.y - other_point.y)**2)
````

````python
# We can see that we can define x, and y while creating the object, 
# Then we can use our methods to apply any process we want on them.

point4 = Point(3, 4)
print(point4.x, point4.y)

point4.reset()
print(point4.x, point4.y)

# try not iinitalizing the argument to see the defualt
point5= Point()
print('point5: \t', point5.x, point5.y)
````

## Who can access my data?


**In  Most object-oriented programming languages (not python):**
*   Some attributes and methods on an object are marked private, meaning only that object can access them

*   Others are marked protected, meaning only that class and any subclasses have access.
*   The rest are public, meaning any other object is allowed to access them.

---

**But In python:**


*   all methods and attributes on a class are publicly available
*   We can prefix an attribute or method with an underscore to indicate that it's an internal variable
* But there is nothing inside the interpreter to stop us from accessing it
* We may not have any idea what future uses our classes may be put to.


---

**WE can prefix attribute/ method with a double underscore: this will perform name mangling**

This basically means that the method can still be called by outside objects if they really want to do it, but it requires extra work and is a strong indicator that you demand that your attribute remains private.



````python
class SecretString:
    '''A not-at-all secure way to store a secret string.'''
    
    def __init__(self, plain_string, pass_phrase):
        self.__plain_string = plain_string
        self.__pass_phrase = pass_phrase

    def decrypt(self, pass_phrase):
        '''Only show the string if the pass_phrase is correct.'''
        if pass_phrase == self.__pass_phrase:
            return self.__plain_string
        else:
            return ''
````

````python
secret_string = SecretString("ACME: Top Secret", "antwerp")
print(secret_string.decrypt("antwerp"))
````

**But we will face an AttributeError if we tried to access the attribute directly**

````python
secret_string.__plain_string
````

It looks like it works and we could protect the atrribute, but in fact we can access atrributes with double underscore perfix using (_classname) before the attribute name as follow:

````python
secret_string._SecretString__pass_phrase
````

````python
secret_string._SecretString__plain_string
````

````python
import datetime

# We will use last_id to give unique id for each Note object
last_id = 0         
class Note:
    
    def __init__(self, memo, creation_date= datetime.date.today(), tag= ''):
        
        self.memo = memo
        self.creation_date = creation_date
        self.tag = tag
        # call last_id using global to access it from outside the class
        global last_id
        last_id +=  1
        self.id = last_id
        
    def match_memo(self, memo_filter):
        # return True if memo_filter text is a part of memo text
        return memo_filter in self.memo
    
    def match_tag(self, tag_filter):
        # return True if tag_filter text is a part of tag text
        return tag_filter in self.tag


class Notebook:

    def __init__(self):
        self.notes = []
        
    def new_note(self, memo, creation_date= datetime.date.today(), tag= ''):
        ''' This is a Composition relation, we create a new Note object
            every time we call this method '''
        note = Note(memo, creation_date , tag)
        self.notes.append(note)
    
    # perfix method/ attribute with underscore mean it's for internal use only
    def _find_note(self, note_id):
        #find note object with id = note_id
        for note in self.notes:
            if note.id == note_id:
                return note
        return None

    def modify_memo(self, note_id, memo):
        self._find_note(note_id).memo = memo 

    def modify_tags(self, note_id, tag):
        self._find_note(note_id).tag = tag 

    def modify_creation_date(self, note_id, creation_date):
        self._find_note(note_id).creation_date = creation_date
        
    def search(self, filter ):
        return [note for note in self.notes if note.match_memo(filter)]
````

We have built till now a Notebook calss that has a composition relation with note class, Which mean that notebook object can include infinite numbers of note objects (one to many relation).

This is exactly what it had been planned in design stage, which also make sense, any notebook have multible notes. 

We will now build an object manager, which will automatically create a Notebook object, and give the user an easy manner to access it to:

- show_notes
- search_notes
- add_note
- modify_note
- quit

````python
import sys

class menu:
    def __init__(self):
        self.notebook = Notebook()
        self.choices = {    "1": self.show_notes,
                            "2": self.search_notes,
                            "3": self.add_note,
                            "4": self.modify_note,
                            "5": self.quit
                        }

    def display_menu(self):
        print("""
                1: show_notes,
                2: search_notes,
                3: add_note,
                4: modify_note,
                5: quit
                """)

    def run(self):
        # Display the menu and respond to choices.
        while True:
            self.display_menu()
            choice = input("Enter an option: ")
            # use choices argument to retrive method
            action = self.choices.get(choice)
            if action:
                action()
            else:
                print("{0} is not a valid choice".format(choice))

    def show_notes(self, notes = None):
        # disply notes content, if None, display content of all existed notes
        if not notes:
            notes = self.notebook.notes
        for note in notes:
            print("{0}:  {1} \n        creation_date:  {2} \t\t\t tag:  {3} \n".format(
                note.id, note.memo, note.creation_date, note.tag))

    def search_notes(self):
        # display all notes that includes the filter
        filter = input("enter the words that you want to search for: ")
        notes =  self.notebook.search(filter)
        if notes:
            self.show_notes(notes)
        # show_notes return all notes if it took None arguments, so we  make sure that won't happen
        else:
            print("There is no note with these words")

    def add_note(self):
        # add new notr yo our menu
        memo = input("input your  note:  ")
        creation_date = input("input the creation date, press enter to use today date:  ")
        tag = input("input the tag, press enter to leave it blank:  ")
        if not creation_date:
            creation_date= datetime.date.today()
        self.notebook.new_note(memo, creation_date, tag)

    def modify_note(self):
        # modify memo/ creation_date/ tag using  memo id
        id = int(input("input your  note id:  "))
        new_memo = input("input new memo:  ")
        new_creation_date = input("input new creation date:  ")
        new_tag = input("input new tag:  ")
        if  new_memo:
            modify_memo(id, new_memo)
        if new_creation_date:
            modify_creation_date(id, new_creation_date)
        if new_tag:
            modify_tag(id, new_tag)

    def quit(self):
        print("closing the notebook for today")
        sys.exit(0)
````
